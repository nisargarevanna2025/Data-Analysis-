```python
df = df.dropna(subset=['Amount'])
```

```python
df.nunique()
```

```python
# Remove duplicate rows
df.drop_duplicates(inplace=True)

# Drop the 'Unnamed: 22' column due to mixed types and many missing values
if 'Unnamed: 22' in df.columns:
    df = df.drop('Unnamed: 22', axis=1)

# Sort the DataFrame by 'Order ID'
df.sort_values(by='Order ID', inplace=True)

# Display the cleaned and sorted DataFrame
display(df.head())
```

```python
# Handle remaining null values
# For categorical columns with nulls, we can fill with a placeholder like 'Unknown' or the mode
for col in ['Courier Status', 'currency', 'ship-city', 'ship-state', 'ship-country', 'promotion-ids', 'fulfilled-by']:
    if col in df.columns:
        df[col] = df[col].fillna('Unknown')

# For numerical columns with nulls (ship-postal-code), we can fill with a placeholder or the mean/median
if 'ship-postal-code' in df.columns:
    df['ship-postal-code'] = df['ship-postal-code'].fillna(-1) # Using -1 as a placeholder for missing postal codes

# Drop columns that are not useful or have too many missing values
if 'Unnamed: 22' in df.columns:
    df = df.drop('Unnamed: 22', axis=1)

# Ensure 'Date' column is in datetime format, handling mixed formats
df['Date'] = pd.to_datetime(df['Date'], format='mixed', dayfirst=False)

# Sort the DataFrame by 'Order ID' (already done in previous step, but good to ensure)
df = df.sort_values(by='Order ID')

display(df.isnull().sum())
display(df.head())
```

```python
display(df.isnull().sum().sort_values(ascending=False))
```

```python
# Ensure the DataFrame is sorted by 'Order ID'
df = df.sort_values(by='Order ID')

# Display the head of the sorted DataFrame
display(df.head())
```

# Task
Generate Python code to create visualizations for sales data. The visualizations should include a bar plot of sales by category, a time series plot of sales over time, and a bar plot of sales by fulfillment type.

## Visualize sales by category

### Subtask:
Create a bar plot to show the total sales amount for each product category.


**Reasoning**:
Group the data by category and calculate the sum of the amount, then create a bar plot to visualize the total sales amount by category as requested in the instructions.



```python
category_sales = df.groupby('Category')['Amount'].sum().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
sns.barplot(x=category_sales.index, y=category_sales.values)
plt.title('Total Sales Amount by Category')
plt.xlabel('Category')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## Visualize sales over time

### Subtask:
Create a time series plot to show the total sales amount over time using the 'Date' column.


**Reasoning**:
Group the data by date and sum the amount, then create a time series plot of the total sales over time.



```python
daily_sales = df.groupby('Date')['Amount'].sum()

plt.figure(figsize=(15, 7))
plt.plot(daily_sales.index, daily_sales.values)
plt.title('Total Sales Amount Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Visualize sales by fulfillment type

### Subtask:
Create a bar plot to compare the total sales amount for different fulfillment types (Amazon vs Merchant).


**Reasoning**:
Group the DataFrame by 'Fulfilment' and calculate the sum of 'Amount' to get sales per fulfillment type, then create and display a bar plot to visualize this data as instructed.



```python
fulfillment_sales = df.groupby('Fulfilment')['Amount'].sum()

plt.figure(figsize=(8, 6))
sns.barplot(x=fulfillment_sales.index, y=fulfillment_sales.values)
plt.title('Total Sales Amount by Fulfillment Type')
plt.xlabel('Fulfilment Type')
plt.ylabel('Total Sales Amount')
plt.show()
```
